```python
import csv
import numpy as np
import pandas as pd
import tkinter as tk
import seaborn as sns
import matplotlib.pyplot as plt
from tkinter import * 
from tkinter import ttk
from tkinter import messagebox
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix as cm
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import classification_report, confusion_matrix
```

Nümerik hale getirilen veri bir dizi işlemden geçirildi ve değişkenlere yazdırıldı

```python
fileName = "datasetDuzenlenmisHali_Numerik_5000.csv"
#Csv dosyası fileName değişkenine atandı
dataset = pd.read_csv(fileName, sep = ';')
#fileName değişkeni pandas kütüphanesi yardımıyla okundu, sep ile ayrılarak sütun sütun olmak üzere dataset değişkenine atandı
X = dataset.iloc[:, :-1].values
#Csv dosyasında bulunan verilerin son sütun dışındakilerini alır
y = dataset.iloc[:, 7].values
#Csv dosyasında bulunan son sütun verisini alır
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
#Verileri %20'lik ve %80'lik kısım olmak üzere iki alt gruba ayırır."train_test_split" sayesinde manuel olarak bölmemiz gerekmez
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X_train, y_train)
#KNN modeli oluşturuldu
y_pred = classifier.predict(X_test)
#Tahmin
cm1 = cm(y_test, y_pred)
#cm1 = Confusion Matrix
#accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
mae = mean_absolute_error(classifier.predict(X_test), y_test)
#mae = Ortalama Mutlak Hata
mse = mean_squared_error(classifier.predict(X_test), y_test)
#mse = Ortalama Kare Hata
rmse = np.sqrt(mse)
#rmse = Kök Ortalama Kare Hata 
kappa = cohen_kappa_score(y_test, y_pred)
#kappa = Cohen'in Kappa Katsayısı
matthews = matthews_corrcoef(y_test, y_pred)
#matthews = Matthews korelasyon katsayısı
```
